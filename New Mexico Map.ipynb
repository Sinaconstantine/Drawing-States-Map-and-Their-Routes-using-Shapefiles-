```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from dbfread import DBF
import numpy as np
import shapefile
```

```python
sf=shapefile.Reader('tl_2019_35_prisecroads.shp')
shape_ex = sf.shape()
```

```python
map = plt.figure(figsize=(8, 8), dpi=300)

map = Basemap(width=1100000,height=1200000,
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',area_thresh=1000.,projection='lcc',\
            lat_1=-0.3,lat_2=0.2,lat_0=34.5,lon_0=-105.8)
x_lon = np.zeros((len(shape_ex.points),1))
y_lat = np.zeros((len(shape_ex.points),1))
for shape in list(sf.iterShapes()):
    x_lon = np.zeros((len(shape.points),1))
    y_lat = np.zeros((len(shape.points),1))
    for ip in range(len(shape.points)):
        x_lon[ip] = shape.points[ip][0]
        y_lat[ip] = shape.points[ip][1]
    x, y = map(x_lon, y_lat)
    plt.plot(x, y,'b',linewidth=0.5,alpha=0.1)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='white', lake_color='aqua')

map.drawcountries()
map.drawstates(color='0.5')

plt.title('New Mexico, Primary and Secondary Roads')

plt.axes()
plt.show()
```
